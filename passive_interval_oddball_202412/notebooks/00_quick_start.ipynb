# Quick start

Use this notebook to confirm the project layout, find packed sessions, and copy the exact command for a first report run. It does not load the large imaging arrays.

## 1. Start from the project root

Open this notebook from `passive_interval_oddball_202412/notebooks`. The next cell moves Python to the project root so imports and relative paths match `main.py`.

```python
from pathlib import Path
import os

repo = Path.cwd()
if repo.name == 'notebooks':
    repo = repo.parent
os.chdir(repo)
print(repo)
```

## 2. Check the expected folders

`results_pack` is the fastest path because it contains compact per-session `.h5` files. Use raw `results` only when you need to run trialization from Suite2p, voltage, Bpod, and camera files.

```python
for folder in ['results_pack', 'results', 'modules', 'plot', 'modeling', 'webpage']:
    path = repo / folder
    print(f'{folder:12s}', 'OK' if path.exists() else 'missing')

packed_h5 = sorted((repo / 'results_pack').glob('*/*/*.h5'))
archives = sorted((repo / 'results_pack').glob('*.7z'))
print(f'packed h5 sessions: {len(packed_h5)}')
print(f'packed archives: {len(archives)}')
print('first sessions:')
for path in packed_h5[:5]:
    print(' ', path.relative_to(repo))
```

## 3. Pick a subject/config

The command-line names are `YH01VT`, `YH02VT`, `YH03VT`, `YH14SC`, `YH16SC`, `YH17VT`, `YH18VT`, `YH19VT`, `YH20SC`, `YH21SC`, `PPC`, and `V1`.

```python
from session_configs import all_config_list

names = ['YH01VT', 'YH02VT', 'YH03VT', 'YH14SC', 'YH16SC',
         'YH17VT', 'YH18VT', 'YH19VT', 'YH20SC', 'YH21SC', 'PPC', 'V1']

for name, cfg in zip(names, all_config_list):
    n_sessions = sum(len(sc['list_session_name']) for sc in cfg['list_config'])
    print(f'{name:6s} -> {cfg["output_filename"]:18s}  {n_sessions:3d} sessions')
```

## 4. First report command

Run one packed subject first. It skips trialization and writes `results/<output_filename>.html`.

```python
SUBJECT = 'YH21SC'
print(f'python main.py --config_list {SUBJECT} --use_packed')
```

Leave `RUN_REPORT = False` while exploring. Set it to `True` only when you want the full report generation to run from this notebook.
